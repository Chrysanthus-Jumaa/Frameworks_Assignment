```python
# DELETING THE COLUMNS OF NO SIGNIFICANCE
new_ds = dataset.drop(['sha', 'doi','mag_id','who_covidence_id','arxiv_id','s2_id','pdf_json_files','pmc_json_files'], axis=1)
```

```python
# DELETING ROWS WITH NULL VALUES IN SOME COLUMNS
new_ds = new_ds.dropna(subset=['url', 'title','authors','journal','pmcid','pubmed_id'])
```

```python
# CARRYING A WORD COUNT OF THE CHARACTERS IN EACH FIELD OF THE 'abstract' column
new_ds.loc[:, 'abstract_word_count'] = new_ds['abstract'].fillna('').apply(lambda x: len(x.split()))
```

```python
new_ds
```

```python
# CONVERTING MY DATES TO DATETIME FORMATS
new_ds.loc[:, 'publish_time'] = pd.to_datetime(new_ds['publish_time'], errors='coerce')

# EXTRACTING YEAR AND STORING IT IN A NEW COLUMN
new_ds.loc[:, 'publish_year'] = new_ds['publish_time'].dt.year
```

```python
new_ds
```

```python
# VIEWING NULL VALUES PER COLUMN
null_count4 = new_ds.isnull().sum()
print(null_count4)
```

```python
new_ds = new_ds.dropna(subset=['publish_time','publish_year'])
new_ds
```

```python
# VIEWING NULL VALUES PER COLUMN
null_count5 = new_ds.isnull().sum()
print(null_count5)
```

```python
unique_years = sorted(new_ds['publish_year'].dropna().unique().tolist())
print(unique_years)
```

```python
# PLOTTING A LINE GRAPH OF PUBLICATIONS OVER TIME
# Group by year and count publications
year_counts = new_ds['publish_year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-', color='teal')
plt.title('Number of Publications per Year')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Get distinct sources
distinct_sources = sorted(new_ds['source_x'].dropna().unique().tolist())

# Display the number of these sources
print(len(distinct_sources))
```

```python
# Count papers by source
source_counts = new_ds['source_x'].value_counts()

# Plot
plt.figure(figsize=(14, 6))
source_counts.plot(kind='bar', color='mediumseagreen')
plt.title('Distribution of Papers by Source')
plt.xlabel('Source')
plt.ylabel('Number of Papers')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# GENERATING A WORD CLOUD OF PAPER TITLES
!pip install wordcloud
from wordcloud import WordCloud

# Combine all titles into one string
text = ' '.join(new_ds['title'].dropna().tolist())

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Paper Titles')
plt.tight_layout()
plt.show()
```

```python
# Count publications by journal
journal_counts = new_ds['journal'].value_counts().head(20)  # Top 20 journals

# Plot
plt.figure(figsize=(14, 6))
journal_counts.plot(kind='bar', color='darkorange')
plt.title('Top Publishing Journals')
plt.xlabel('Journal')
plt.ylabel('Number of Publications')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
